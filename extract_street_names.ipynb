```python
import osmnx as ox
import pandas as pd
import geopandas as gpd
import networkx as nx
import random
import string
```

```python
# Get list of all street names in Paris (includes duplicates)
place='Paris, France'
language='FRA'

G = ox.graph_from_place(place, network_type='all', which_result=1) 

edge_attributes = ox.graph_to_gdfs(G, nodes=False)

street_names = [x for x in list(edge_attributes["name"]) if str(x) != 'nan']
```

```python
search_phrases =[]
for street_name in street_names:
    if type(street_name) is str:
        street_name = street_name.replace(',', '')
        street_name = street_name.replace('…', '')
        search_phrases.append(street_name)

search_phrases = list(dict.fromkeys(search_phrases))

with open('combined.txt', encoding = "utf-8") as f:
    for line in f:
        for phrase in search_phrases:
            if phrase.lower() in line.lower():
                print(phrase)
```

```python
# SORTED VERSION

# open the input file
with open('combined_street_names.txt', 'r') as f:
    # create a dictionary to store the counts of each line
    line_counts = {}

    # iterate over each line in the file
    for line in f:
        # remove leading and trailing whitespace from the line
        line = line.strip()

        # if the line is not in the dictionary, add it and set the count to 1
        if line not in line_counts:
            line_counts[line] = 1
        # if the line is already in the dictionary, increment the count
        else:
            line_counts[line] += 1

# open the output file
with open('street_names_count.txt', 'w') as f:
    # sort the lines by count in descending order
    sorted_lines = sorted(line_counts.items(), key=lambda x: x[1], reverse=True)

    # write each line and its count to the output file
    for line, count in sorted_lines:
        f.write(f"{line}, {count}\n")
```
